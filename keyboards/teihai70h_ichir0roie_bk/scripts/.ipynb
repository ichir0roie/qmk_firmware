```python
import json
kmpkg_path = "kmpkg.json"

with open(kmpkg_path, "r", encoding="utf-8")as f:
    kmpkg = json.load(f)
kmpkg
```

```python
lo = kmpkg["layouts"][0]["data"]["keyEntities"]
lo
layout = []
for entity in lo:
    x = entity["x"]
    y = entity["y"]
    line = {
        "matrix": [x, y],
        "x": x,
        "y": y
    }
    layout.append(line)
layout
```
