```python
import shutil
from glob import glob
```

```python
kbs = glob("keyboards/*", recursive=True)
kbs
```

```python
del_kbs = []

my_keyborads = [
    "ergoarrows",
    "ergodash",
    "nora",
    "teihai70h_ichir0roie"
]

for kb in kbs:
    if kb.split("\\")[1] in my_keyborads:
        print(kb)
        continue
    del_kbs.append(kb)
```

```python
for kb in del_kbs:
    if "." in kb:
        continue
    shutil.rmtree(kb)
```
